```python
import numpy as np
import pandas as pd
df = pd.read_csv('weather.csv')
df
```

```python
df.groupby(['Station.City'])['Data.Temperature.Max Temp'].mean()
```

```python
df.isnull()
```

```python
df['Date.Full'].ffill()
```

```python
print(df['Date.Full'].dtype)
df['Date.Full'] = pd.to_datetime(df['Date.Full'])
print(df['Date.Full'].dtype)
```

```python
result = df.groupby(['Station.City','Station.Code']).agg({'Data.Temperature.Avg Temp':sum})
result.sort_values(by='Data.Temperature.Avg Temp')
```

```python
temp = pd.cut(df['Data.Wind.Speed'],3) ## 3 is no of bins
print(temp)
```

```python
group = df.groupby(['Station.State',temp])
print(group.size().unstack())
```
